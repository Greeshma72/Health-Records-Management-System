```python
#Converting categorical variables i.e, Gender, Blood Type, Medical Condition, InsuranceProvider
from sklearn.preprocessing import LabelEncoder

# List of columns to be label encoded
label_columns = ['Gender', 'Blood Type', 'Medical Condition', 'Insurance Provider']

# Initialize label encoder
label_encoder = LabelEncoder()

# Apply label encoding to each column
for column in label_columns:
    df1[f'{column}_Label'] = label_encoder.fit_transform(df1[column])

print("\nLabel Encoded DataFrame:")
print(df1)
```

```python
from scipy.stats import f_oneway, ttest_ind
# Check for any unexpected values in 'Medical Condition' and 'Insurance Provider' columns
print("Unique Medical Conditions:", df['Medical Condition'].unique())

# Convert age and billing amount columns to NumPy arrays
age = df['Age'].values
billing_amount = df['Billing Amount'].values

# Construct groups for statistical tests
groups_age = [age[df['Medical Condition'] == condition] for condition in df['Medical Condition'].unique()]
groups_billing = [billing_amount[df['Medical Condition'] == provider] for provider in df['Medical Condition'].unique()]

# Perform statistical tests
f_statistic_age, p_value_age = f_oneway(*groups_age)

# Print results
print("\nANOVA Test Results - Age across different medical conditions:")
print("F-statistic:", f_statistic_age)
print("p-value:", p_value_age)

f_statistic_billing, p_value_billing = f_oneway(*groups_billing)

print("\nIndependent f-test Results - Billing Amount across different Medical Conditions:")
print("f-statistic:", f_statistic_billing)
print("p-value:", p_value_billing)
```

Patient Segmentation

```python
df3 = df.copy()
df3['Date of Admission'] = pd.to_datetime(df3['Date of Admission'])
df3['Discharge Date'] = pd.to_datetime(df3['Discharge Date'])

# Calculate Length of Stay (in days)
df3['Length of Stay'] = (df3['Discharge Date'] - df3['Date of Admission']).dt.days

# Define a function to categorize patients based on 'Admission Type'
def categorize(row):
    if row['Admission Type'] == 'Urgent':
        return 'High Priority Patients'
    elif row['Admission Type'] == 'Emergency':
        return 'Critical Patients'
    else:
        return 'Elective Patients'

# Apply the function to create a new 'category' column
df3['category'] = df3.apply(categorize, axis=1)

# Calculate summary statistics for each category
summary = df3.groupby('category').agg({
    'Age': 'mean',
    'Billing Amount': 'mean',
    'Length of Stay': 'mean'
}).rename(columns={
    'Age': 'Average Age',
    'Billing Amount': 'Average Billing Amount',
    'Length of Stay': 'Average Length of Stay',
    'Patient ID': 'Number of Patients'
})

print("Summary Statistics for Each Category:")
print(summary)
```

```python
# Plotting Box Plots
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.boxplot(x='category', y='Age', data=df3)
plt.title('Box Plot of Age by Category')
plt.xlabel('Category')
plt.ylabel('Age')

plt.subplot(1, 3, 3)
sns.boxplot(x='category', y='Length of Stay', data=df3)
plt.title('Box Plot of Length of Stay by Category')
plt.xlabel('Category')
plt.ylabel('Length of Stay (days)')

plt.tight_layout()
plt.show()

# Plotting Violin Plots
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 2)
sns.violinplot(x='category', y='Billing Amount', data=df3)
plt.title('Violin Plot of Billing Amount by Category')
plt.xlabel('Category')
plt.ylabel('Billing Amount')

plt.tight_layout()
plt.show()
```

Calculating Percentages of Test Results

```python
# Calculate percentages of test results for each medical condition
test_result_percentages = df3.groupby(['Medical Condition', 'Test Results']).size().unstack(fill_value=0)
test_result_percentages = test_result_percentages.apply(lambda x: x / x.sum() * 100, axis=1)

print(test_result_percentages)
```

```python
#Visualizing Test Result Composition with Pie Charts/Donut Charts
# Function to plot pie chart
def plot_pie_chart(condition, data):
    labels = data.columns
    sizes = data.loc[condition]
    colors = sns.color_palette('Set1', len(labels))
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
    plt.title(f'Test Result Composition for {condition}')
    plt.show()

# Plot pie charts for selected medical conditions
selected_conditions = test_result_percentages.index[:3]  # Select first 3 conditions for example
for condition in selected_conditions:
    plot_pie_chart(condition, test_result_percentages)
```
